###Data Retrieval

Shows the name of the song with most and minimum popularity

```python
row_with_max_popularity = df[df['popularity'] == df['popularity'].max()]
p=row_with_max_popularity['name']
print("The most popular song is", p)

row_with_min_popularity = df[df['popularity'] == df['popularity'].min()]
r=row_with_min_popularity['name']
print("The least popular song is",r)
```

The number of songs in each album

```python
df['album'].value_counts()
```

Mean Maximum and Minimum of danceability of a song

```python
aggregated_data = df.groupby('name')['danceability'].agg(['mean', 'max', 'min'])
print(aggregated_data.tail())
```

Average energy of each album

```python
avg_energy_by_album = df.groupby('album')['energy'].mean().sort_values(ascending=False)
print(avg_energy_by_album)
```

The Longest and The Shortest tracks

```python
longest_tracks = df.sort_values(by='duration_ms', ascending=False).head(1)
shortest_tracks = df.sort_values(by='duration_ms').head(1)
print(longest_tracks['name'])
print(shortest_tracks['name'])
```

## Data Visualisation

```python
Album= df['album']
Name= df['name']
Albums=np.unique(Album)
album1=str(Albums)
#print(album1)
```

```python
wc = WordCloud(background_color="white").generate(album1)
plt.figure(figsize=(10,10))
plt.imshow(wc)
plt.axis("off");
```

```python
df["release_date"] = pd.to_datetime(df["release_date"])
df["year"] = df["release_date"].dt.year
fig=px.area(df.groupby('year',as_index=False).count().sort_values(by='name',ascending=False).sort_values(by='year'),x='year',y='name',markers=True,labels={'name':'Total songs'},color_discrete_sequence=['green'],title='<b>Year by Year Songs collection')
fig.update_layout(hovermode='x',title_x=0.5)
```

This Heat Map is used to visualize and display a geographic distribution of variables across the dataset..

Correlations in the following heat map

1. There is a strong positive link between loudness and energy (0.80).

2. There is a strong negative link between loudness/energy and acousticness (-0.72 & -0.73)

3. Track number on an album has little to no impact on all other features.

```python
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
corr = df.corr()
sns.heatmap(corr,annot=True,cmap="crest")
```

Plotly Express is a package that allows to show interactive graphs.
With px. line , each data point is represented as a vertex

```python
fig=px.line(df.sort_values(by='popularity',ascending=False).head(25),x='name',y='popularity',hover_data=['album'],color_discrete_sequence=['purple'],markers=True,title='<b> Popularity of songs')
fig.show()
```
